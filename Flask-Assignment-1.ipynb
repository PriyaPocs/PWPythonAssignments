Q1. What is Flask Framework? What are the advantages of Flask Framework?

Answer(Q1):

Flask is a lightweight Web Server Gateway Interface (WSGI) web application framework.

A micro web framework that has minimal dependencies on external libraries, written in Python, which was formed for a faster and easier use, and also has the ability to scale up to complex applications.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.


```python
# Answer(Q2)
# python Flask App code
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"  

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

Answer(Q3):

App Routing means mapping the URLs to a specific function that will handle the logic for that URL. 

Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler.
