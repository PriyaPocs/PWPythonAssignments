Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Answer(Q2):


The scatter diagram graphs pairs of numerical data, with one variable on each axis, to look for a relationship between them. 
It's a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data usually marked by points on the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize=(6,4))
plt.scatter(x,y , c = 'g',alpha=.9,marker='+')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title("Graph displaying x vs y")
plt.grid()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


Answer(Q3):

subplot() function adds subplot to a current figure at the specified grid position. It adds one subplot at a time. 


```python
# importing required libraries
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


fig, axs = plt.subplots(2, 2)
plt.figure(figsize=(6,4))

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 1].plot(x2, y2, 'tab:orange')
axs[0, 1].set_title('Line 2')
axs[1, 0].plot(x3, y3, 'tab:green')
axs[1, 0].set_title('Line 3')
axs[1, 1].plot(x4, y4, 'tab:red')
axs[1, 1].set_title('Line 4')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Answer(Q4):

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. 



```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

fig = plt.figure(figsize = (6, 3))
 
# creating the bar plot
plt.bar(company, profit, color ='maroon',
        width = 0.4)
 
plt.xlabel("companies")
plt.ylabel("profit")
plt.title("Companies Vs Profits details")
plt.show()
```

```python
fig = plt.figure(figsize = (6, 3))
 
# creating the bar plot
plt.barh(company, profit, color ='maroon')
plt.xlabel("profit")
plt.ylabel("companies")
plt.title("Horizondal Bar Graph")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Answer(Q5):

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.


```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
 
fig = plt.figure(figsize =(6, 5))
 
# Creating axes instance
ax = fig.add_axes([0,0,1,1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```
