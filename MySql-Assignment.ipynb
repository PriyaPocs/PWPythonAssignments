Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.


Answer(Q2):

A data definition language (DDL) is a computer language used to create and modify the structure of database objects in a database. These database

objects include views, schemas, tables, indexes, etc.

The CREATE TABLE command creates a new table in the database.  

1.CREATE TABLE Persons (
    PersonID int,
    LastName varchar(255),
    FirstName varchar(255),
    Address varchar(255),
    City varchar(255)
);

2.The ALTER TABLE command adds, deletes, or modifies columns in a table.

The ALTER TABLE command also adds and deletes various constraints in a table.

ALTER TABLE Persons ADD Email varchar(255);

3.The TRUNCATE TABLE command deletes the data inside a table, but not the table itself.

TRUNCATE TABLE Persons;

4.The DROP TABLE command deletes a table in the database.

DROP TABLE Persons;


Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

Answer(Q3):
    
The structured query language (SQL) commands deal with the manipulation of data present in the database that belongs to the DML or Data 

Manipulation Language. This includes most of the SQL statements.  

1.INSERT: Insert command is used to insert data into a table.

Syntax:
    Insert into <table_name> (column list) values (column values);

Eg:
    Insert into Employee(Emp_id, Emp_name) values (001, “ bhanu”);
    Insert into Employee(Emp_id, Emp_name) values (002, “ hari”);
    Insert into Employee(Emp_id, Emp_name) values (003, “ bob”);

2.UPDATE: Update command is used to update existing data within a table.

Syntax:
    UPDATE <table_name> SET column_number =value_number WHERE condition;

Eg:
  UPDATE Employee SET Emp_name= Ram WHERE Emp_id= 001;
  
  
3.DELETE: Delete command is used to delete records from a database table.

Syntax:
    Delete from <table_name>WHERE condition;

Eg:
  DELETE from Employee WHERE Emp_id=002;

Q5. Explain Primary Key and Foreign Key.


Answer(Q5):

Primary Key:
  A primary key generally focuses on the uniqueness of the table. It assures the value in the specific column is unique.
  
  CREATE TABLE Customer(
    Customer_id int primary key,
    Customer_name varchar(20),
    Customer_Address varchar(20),
)
  
  
Foreign Key:
  A foreign key is generally used to build a relationship between the two tables. The table allows only one primary key.
  
  CREATE TABLE SALES(
    Customer_id int FOREIGN KEY REFERENCES Customer(Customer_id)
    Item_id int,
    Payment_Mode varchar(20), 
)
